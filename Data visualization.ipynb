```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
import tensorflow as tf
```

Data visualization using seaborn


line plot in jupyter notebook

```python
tips = sns.load_dataset("tips")
```

Scatter plot

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', style='sex', size='size')
plt.title("scatter plot of Total Bill vs Tip")
plt.show()
```

Bar Plot

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.title("Average Tip by Day and Gender")
plt.show()
            
```

Box plot

#explore the distribution of total_bill across different days

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x='day', y='total_bill', hue='time')
plt.title("Box Plot of Total Bill by Day and Time")
palette='set2'
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=tips, x='sex', y='tip', hue='smoker', split=True)
plt.title("violin plot of tips by gender and smoking status")
plt.show()
```
